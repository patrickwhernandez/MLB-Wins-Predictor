### Import Packages

```python
import pandas as pd
import numpy as np
```

### Load in Data

```python
pd.set_option('display.max_columns', 120)
pd.set_option('display.max_rows', 120)
```

```python
df = pd.read_csv("mlb_season_stats.csv", index_col=[0]).reset_index(drop=True)
df.head()
```

### Clean Data

```python
# Remove Win-Probability Stats
df.drop(list(df.filter(regex="WPA")), axis=1, inplace=True)
df.drop(list(df.filter(regex="RE24")), axis=1, inplace=True)
```

```python
# Remove repeating data (Team, Games Played)
df.drop(list(df.filter(regex="Tm.")), axis=1, inplace=True)
df.drop(["G.1", "GS", "GF"], axis=1, inplace=True, errors="ignore")
df.columns = df.columns.str.replace(".1", " (P)")
```

```python
# Remove 2020 season data (shortened season)
df = df[df["Year"] != 2020]
```

```python
df.dtypes
```

```python
# Convert percentages to float datatype
for column in df.columns:
    if column != "Tm":
        if df[column].dtype == "object":
            df[column] = df[column].str.rstrip('%').astype(np.float64) / 100
```

### Create new file

```python
df.to_csv("mlb_season_stats_clean.csv")
```
